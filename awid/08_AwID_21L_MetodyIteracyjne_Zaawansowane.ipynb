### Metoda najszybszego spadku

```julia
function steepdes_positive(A,b, x0)
    xk = x0;
    rk = 1;
    maxiter = 100;
    iter = 1;
    R = []
    Rk = []
    X = []
    push!(X, xk)
    while (norm(rk) > 1e-5 && iter < maxiter)
        rk = b - A*xk;
        push!(Rk, rk)
        ak = rk'*rk / (rk'*A*rk) ;      
        xk = xk + ak*rk        
        iter = iter + 1;
        push!(R, norm(rk))
        println(xk)
        push!(X, xk)
    end
    
    return xk, R, X, Rk
end


```

Metoda najmniejszego residuum
Stosujemy gdy macierz jest nieosobliwa (niesymetryczna), i przyjmujemy

$$V_k = r_k = b-Ax_k, \quad U_k = Ar_k$$

wówczas:

$$r_k^TA^TAr_ka_k = A^Tr_k^Tr_k \quad \rightarrow \quad a_k = \frac{r_k^TA^Tr_k}{r_k^TA^TAr_k}$$

naiwny kod w

```julia
function steepdes_nonsingular(A,b, x0)
    xk = x0;
    rk = 1;
    maxiter = 10000;
    iter = 1;
    R = []
    while (norm(rk) > 1e-5 && iter < maxiter)
        rk = b - A*xk;
        # ak = rk'*A'*rk / (rk'*A'*A*rk) ;   
        # lepiej
        ak = rk'*A'*rk / ((A*rk)'* (A*rk)) ;   
        xk = xk + ak*rk        
        iter = iter + 1;
        push!(R, norm(rk))
    end
    return xk, R
end
```

```julia
A = [10 2 4
     2 50 5
     4 5 -30];
b = [1;1;1];

x, R = steepdes_nonsingular(A,b,[1,1,0])
#x, R, X, Rk = steepdes_positive(A,b,[1;1;0])
```
